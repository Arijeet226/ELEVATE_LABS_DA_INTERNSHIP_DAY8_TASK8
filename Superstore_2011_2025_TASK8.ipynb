## EDA

### Q1. INITIAL FINDING FOR PRODUCT CATEGORY

```python
category_profit=data.groupby('Category')['Profit'].sum()
category_profit.plot(kind='bar')
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.show()
```

Technology is by far has the most profit.

### Q2. INITIAL FINDING FOR REGION

```python
region_sales=data.groupby('Region')['Sales'].sum()
region_sales.plot(kind="bar")
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

Central region has more sale compared to any other region followed by south and then north.

### Q3.INITIAL FINDINGS FOR MONTHS

```python
data['Order Month']=pd.DatetimeIndex(data['Order Date']).month
month_sales=data.groupby('Order Month')['Sales'].sum()
month_sales.plot(kind='line')
plt.title("Total Sales by Month")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

November has the highest volume of sales.
